```python
import numpy as np
import pandas as pd 
```

```python
news_df = pd.read_csv("investing_com.csv")
news_df
```

```python
news_df=news_df.drop(columns=["headline", "id", "date"])
news_df
```

```python
news_df=news_df.rename(columns={"content":"news_body"})
```

```python
news_df
```

```python
remove_newline = lambda x: x.replace('\n', '')
news_df['news_body'] = news_df['news_body'].apply(remove_newline)
news_df
```

```python
# news_df= news_df.dropna()
news_df = news_df.dropna(subset=['news_body'])
# news_df.to_csv("clean_news_dataset.csv", index=False)
news_df
```

```python
news_df = news_df[news_df['news_body'].apply(lambda x: len(str(x)) > 0)]
news_df
```

```python
news_df_400 = news_df[:400]
news_df_400
```

```python
news_df_1 = news_df_400[:100]
news_df_2 = news_df_400[101:200]
news_df_3 = news_df_400[201:300]
news_df_4 = news_df_400[301:400]
news_df_1
```

```python
news_df_1.to_csv("clean_news_1.csv", index=False)
news_df_2.to_csv("clean_news_2.csv", index=False)
news_df_3.to_csv("clean_news_3.csv", index=False)
news_df_4.to_csv("clean_news_4.csv", index=False)
```
